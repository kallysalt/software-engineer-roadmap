```python
# create a queue (first in first out)

from collections import deque
queue = deque()

print(queue)
```

```python
# add element
# time complexity: O(1)

queue.append(1)
queue.append(2)
queue.append(3)
print(queue)
```

```python
# get the head of the queue (first in)
# time complexity: O(1)

tmp = queue[0]
print(tmp)
```

```python
# remove the head of the queue (first in)
# time complexity: O(1)

tmp = queue.popleft()
print(tmp)
print(queue)
```

```python
# length of queue
# time complexity: O(1)

n = len(queue)
print(n)

# queue is empty?
# time complexity: O(1)

print(len(queue) == 0)
```

```python
# iterate and remove all elements
# time complexity: O(N)

while len(queue) != 0:
    temp = queue.popleft()
    print(temp)
    
```
